Q1 .For the given data points, plot the lines of best fit for dimensionality reduction after applying LDA and PCA.



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

data = np.array([
    [1, 2, 0],
    [3, 5, 0],
    [4, 3, 0],
    [5, 6, 0],
    [7, 5, 0],
    [6, 2, 1],
    [9, 4, 1],
    [10, 1, 1],
    [12, 3, 1],
    [13, 6, 1]
])

X = data[:, :2]
y = data[:, 2]

pca = PCA(n_components=1)
X_pca = pca.fit_transform(X)
line_pca = pca.inverse_transform(X_pca)

lda = LDA(n_components=1)
X_lda = lda.fit_transform(X, y)
lda_direction = lda.scalings_[:, 0]
lda_mean = lda.means_.mean(axis=0)
line_lda = np.array([lda_mean - 5 * lda_direction, lda_mean + 5 * lda_direction])

plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], label='Class 0', color='blue')
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], label='Class 1', color='red')
plt.plot(line_pca[:, 0], line_pca[:, 1], label='PCA Line', color='green')
plt.plot(line_lda[:, 0], line_lda[:, 1], label='LDA Line', color='orange')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('LDA and PCA Lines of Best Fit')
plt.grid()
plt.show()
```

2)Consider the 128- dimensional feature vectors (d=128) given in the “gender.csv” file. (2 classes, male and female)

Dataset Specifications:
Total number of samples = 800
Number of classes = 2 (labeled as “male” and “female”)
Samples from “1 to 400” belongs to class “male”
Samples from “401 to 800” belongs to class “female”
Number of samples per class = 400
Number of dimensions = 128
Use the following information to design classifier:
Number of test cases (first 10 in each class) = 20
Number of training feature vectors ( remaining 390 in each class) = 390
Number of reduced dimensions = d’ (map 128 to d’ features vector)


```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report

data = pd.read_csv('gender.csv')
data_cleaned = data.drop(columns=['Unnamed: 0'])

features = data_cleaned.iloc[:, 1:].astype(float)
labels = data_cleaned['Unnamed: 1']

encoder = LabelEncoder()
labels_encoded = encoder.fit_transform(labels)

features_male, features_female = features[:400], features[400:]
labels_male, labels_female = labels_encoded[:400], labels_encoded[400:]

train_features = pd.concat([features_male[10:], features_female[10:]], ignore_index=True)
train_labels = np.concatenate([labels_male[10:], labels_female[10:]])
test_features = pd.concat([features_male[:10], features_female[:10]], ignore_index=True)
test_labels = np.concatenate([labels_male[:10], labels_female[:10]])

print(train_features.shape, train_labels.shape, test_features.shape, test_labels.shape)

lda_model = LinearDiscriminantAnalysis(n_components=1)
train_features_reduced = lda_model.fit_transform(train_features, train_labels)
test_features_reduced = lda_model.transform(test_features)

nb_classifier = GaussianNB()
nb_classifier.fit(train_features_reduced, train_labels)

predicted_labels = nb_classifier.predict(test_features_reduced)

model_accuracy = accuracy_score(test_labels, predicted_labels)
classification_summary = classification_report(test_labels, predicted_labels, target_names=encoder.classes_)

print(model_accuracy)
print(classification_summary)
```

Fisherfaces- Face classification using LDA (40 classes)
a) Use the following “face.csv” file to classify the faces of 40 different people using LDA.
b) Do not use the in-built function for implementing LDA.
c) Use appropriate classifier taught in class (use any classification algorithm taught in class like Bayes classifier, minimum distance classifier, and so on)
d) Refer to the following link for a description of the dataset: https://towardsdatascience.com/eigenfaces-face-classification-in-python-7b8d2af3d3ea




```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

data = pd.read_csv('face.csv')

features = data.iloc[:, :-1].values
targets = data['target'].values

overall_mean = np.mean(features, axis=0)

unique_classes = np.unique(targets)
num_features = features.shape[1]
scatter_within = np.zeros((num_features, num_features))
scatter_between = np.zeros((num_features, num_features))

for cls in unique_classes:
    class_features = features[targets == cls]
    class_mean = np.mean(class_features, axis=0)
    scatter_within += np.dot((class_features - class_mean).T, (class_features - class_mean))
    class_size = class_features.shape[0]
    mean_diff = (class_mean - overall_mean).reshape(-1, 1)
    scatter_between += class_size * mean_diff.dot(mean_diff.T)

eig_values, eig_vectors = np.linalg.eig(np.linalg.pinv(scatter_within).dot(scatter_between))

sorted_indices = np.argsort(eig_values)[::-1]
eig_vectors = eig_vectors[:, sorted_indices]

k = len(unique_classes) - 1
transformation_matrix = eig_vectors[:, :k]

features_lda = features.dot(transformation_matrix)

X_train, X_test, y_train, y_test = train_test_split(features_lda, targets, test_size=0.3, random_state=42)

class_centroids = {cls: np.mean(X_train[y_train == cls], axis=0) for cls in unique_classes}

predicted_labels = []
for feature in X_test:
    distances = {cls: np.linalg.norm(feature - centroid) for cls, centroid in class_centroids.items()}
    nearest_class = min(distances, key=distances.get)
    predicted_labels.append(nearest_class)

classification_accuracy = accuracy_score(y_test, predicted_labels)
print(f'Classification Accuracy: {classification_accuracy * 100:.2f}%')
```
