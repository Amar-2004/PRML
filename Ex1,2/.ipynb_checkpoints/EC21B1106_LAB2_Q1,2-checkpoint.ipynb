8. Handling Missing Values
a) Check for missing values
b) Fill numeric columns with mean and categorical columns with mode if missing

```python
df.isna()
col_numeric=df.select_dtypes(include=['number']).columns
col_obj=df.select_dtypes(include=['object']).columns
for i in col_numeric:
    df.fillna(df[i].mean(),inplace=True)
for i in col_obj:
    df.fillna(df[i].mode()[0],inplace=True)
df.head()


```

9. Concatenating Data
a) Split the DataFrame into male and female
b) Add 'Group' column to each part
c) Concatenate the two DataFrames

```python
df_male=df.loc[df['Sex']=='M'].copy()
df_female=df.loc[df['Sex']=='F'].copy()
df_male.loc[:,'Gender']='Male'
df_female.loc[:,'Gender']='Female'
df_combined=pd.concat([df_male,df_female])
df_combined.tail(10)
```

 1. Create a figure with 2 rows and 2 columns of subplots. Set the figure size to
 12 inches wide by 10 inches tall

```python
%matplotlib inline
import matplotlib.pyplot as plt
import math
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2,figsize=(12,10))
x=np.linspace(0,10,100)
axes[0,0].plot(x,x,color='blue',linewidth=0.5,label='Y=X')
axes[0,0].plot(x,x**2,color='red',linewidth=1,label='Y=X^2')
axes[0,0].plot(x,x**3,color='green',linewidth=2,label='Y=X^3')
axes[0,0].legend()
axes[0,0].set_title("Basic Line Plots")
y=np.sin(x)
axes[0,1].plot(x,y,linestyle="-",label='solid')
axes[0,1].plot(x,y,linestyle="--",label='dashed')
axes[0,1].plot(x,y,linestyle="-.",label='dash-dot')
axes[0,1].plot(x,y,linestyle=":",label='dotted')
axes[0,1].legend()
axes[0,1].set_title("Linestyles")
y=np.cos(x)
axes[1,0].plot(x,y,label='+',marker='+',markersize=5)
axes[1,0].plot(x,y,label='o',marker='o',markersize=8)
axes[1,0].plot(x,y,label='s',marker='s',markersize=11)
axes[1,0].plot(x,y,label='^',marker='^',markersize=14)
axes[1,0].legend()
axes[1,0].set_title("Markers")
y=np.tan(x)
custom_dash=[2,5,2,5]
line,=axes[1,1].plot(x,y,marker='o',linestyle='-',markersize=8)
#line.set_dashes(custom_dash)
line.set_markerfacecolor('yellow')
line.set_markeredgecolor('red')
for a in axes.flat:
    a.set_xlabel('X-axis')
    a.set_ylabel('Y-axis')
    a.grid(True)
fig.tight_layout()
fig.suptitle("Matplotlib Line Plot Customization")
fig.savefig('matplotlib_assignment.png')
plt.show()

```
