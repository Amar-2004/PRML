Train a single perceptron to learn the two classes in the following table.

```python
d={'Feature1':[2,-1,-1,0,1,-1,1,-1],'Feature2':[2,-3,2,-1,3,-2,-2,-1],'Class':[1,0,1,0,1,0,0,1]}
df=pd.DataFrame(d)
class1=df.loc[df['Class']==1,['Feature1','Feature2']].to_numpy()
class2=df.loc[df['Class']==0,['Feature1','Feature2']].to_numpy()
df['bias']=1
df.loc[df['Class']==0]*=-1
df.drop('Class',axis=1,inplace=True)
table=df.to_numpy()
print(table)
```

```python
alpha=[0.001,0.01,0.5]
iterations=[]
for alpha_value in alpha:
    iteration=0
    weights=np.array([0,0,0])
    error=1
    while error!=0:
        error=0
        for j,i in enumerate(table):
            iteration=iteration+1
            val=weights.dot(i.T)
            if val<=0:
                error=1
                weights=weights+i*(alpha_value)
    iterations.append(iteration)
for i,j in zip(alpha,iterations):
    print(f"for value of alpha={i},it took {j} iterations")

weights=weights/weights[0]
print(f"Final Equation = {weights[0]}x1+{weights[1]}x2+{weights[2]}")
```

```python
#equation of the form y=(-weight[1])x+(-weight[2])
x=np.linspace(-5,5,100)
y=(-weights[1])*x+(-weights[2])
plt.scatter(class1[:,1],class1[:,0],color='yellow')
plt.scatter(class2[:,1],class2[:,0],color='blue')
plt.plot(x,y,color='Green')
plt.xlabel('Feature2')
plt.ylabel('Feature1')
plt.grid(True)
plt.title("Decision Boundary")
```

4.	From the iris dataset, choose the ’petal length’, ’sepal width’ for “setosa” and  “virginica” flowers. Learn a decision boundary for the two features using a single perceptron. Assume that all the weights of the perceptron are initialized as 0 with the learning rate of 0.01. Output the final weight vector and plot the samples and the decision boundary.
[Note: Use iris.csv file in the attachments.]


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df = pd.read_csv('iris.csv')


setosa = df[df['Species'] == 'Iris-setosa'][['PetalLengthCm', 'SepalWidthCm']].to_numpy()
virginica = df[df['Species'] == 'Iris-virginica'][['PetalLengthCm', 'SepalWidthCm']].to_numpy()

X_setosa = np.c_[np.ones(len(setosa)), setosa]  
X_virginica = np.c_[np.ones(len(virginica)), virginica]  


X = np.vstack([X_setosa, X_virginica])
y = np.hstack([np.ones(len(setosa)), -np.ones(len(virginica))])


weights = np.zeros(3)  
learning_rate = 0.01
error = 1
max_iterations = 1000
iteration = 0

while error != 0 and iteration < max_iterations:
    error = 0
    iteration += 1
    
    for i in range(len(X)):
        val = weights.dot(X[i])
        if val * y[i] <= 0:  
            error = 1
            weights = weights + learning_rate * y[i] * X[i]

print(f"Training completed in {iteration} iterations")
print(f"Final weights: w0 (bias) = {weights[0]:.3f}, w1 (petal length) = {weights[1]:.3f}, w2 (sepal width) = {weights[2]:.3f}")


plt.figure(figsize=(10, 6))
plt.scatter(setosa[:, 0], setosa[:, 1], color='blue', label='Setosa')
plt.scatter(virginica[:, 0], virginica[:, 1], color='red', label='Virginica')


x = np.linspace(min(df['PetalLengthCm']), max(df['PetalLengthCm']), 100)
y = (-weights[0] - weights[1]*x) / weights[2]

plt.plot(x, y, 'g-', label='Decision Boundary')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Perceptron Decision Boundary for Iris Setosa vs Virginica')
plt.legend()
plt.grid(True)
plt.show()
```
